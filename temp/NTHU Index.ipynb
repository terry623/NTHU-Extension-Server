```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import requests
import json
```

```python
d = pd.read_pickle('open_course_data_with_syllabus_serialize')
```

```python
def index2elastic(idx, data):
    req = requests.put('http://127.0.0.1:9200/nthu/course/' + str(idx), json=data)
    print(idx, req.json()['result'])
```

```python
for idx, row in d.iterrows():
    index2elastic(idx, row.to_dict())
```
